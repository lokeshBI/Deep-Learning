## ANN Model on Classification Problem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\LOKESH\\Desktop\\python assignment\\Jupyter notebooks for this course\\Refactored_Py_DS_ML_Bootcamp-master\\13-Logistic-Regression\\advertising.csv')
data.head()
```

```python
## after EDA, these are the columns that will help by taking into the data

advertising_data = data[['Daily Time Spent on Site', 'Age', 'Area Income',
       'Daily Internet Usage',  'Male', 'Clicked on Ad']]
advertising_data.head()
```

```python
## split the data into X1 and Y1
X1 = advertising_data.iloc[:, 0:-1].values
y1 = advertising_data.iloc[:, 5].values
print('X1: ','\n',X1,'\n')
print('y1: ','\n',y1)
```

```python
# split the data into train, test sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X1, y1, test_size = 0.2, random_state = 0)
```

```python
# for deeplearnig models scaling is mandatory

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train,'\n')
print(X_test)
```

```python
# importing Keras modules
import keras
from keras.models import Sequential  # for model initialisation
from keras.layers import Dense       # for adding layers to the model
```

```python
# initialising ANN
classifier = Sequential()
```

```python
# adding the input and first hidden layer
# for adding hidden layers --> avg(input variables + output variable)
# in our case here (5+1)/2 = 3

classifier.add(Dense(units = 3, kernel_initializer = 'uniform', activation = 'relu', input_dim = 5))
```

```python
# adding second hidden layer
# for second hidden layer we don't need to add input_dim

classifier.add(Dense(units = 3, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# adding the output layer
# at output we only have 1 node

classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# compiling the ANN

classifier.compile( optimizer= 'adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
```

```python
## fitting the ANN to training data

classifier.fit(X_train, Y_train, batch_size= 10, epochs = 50)
```

```python
# making the predictions

y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
```

```python
# we are getting probabilities, so we have to take cut-off value
y_pred_updated = [  1 if i>0.5 else 0  for i in y_pred]
print(y_pred_updated)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, y_pred_updated)
cm
```
